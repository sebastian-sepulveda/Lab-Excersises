```python
import requests
import json
import os
from PIL import Image

api_key = "c8587e9f54msh872b6008db8c09cp11d6cajsndb39be835356"
api_host = "shazam.p.rapidapi.com"

headers={"X-RapidAPI-Key": api_key, 
         "X-RapidAPI-Host": api_host}

#I figured a lot of this out from the code snippets that are featured on the shazam api site, 
#https://rapidapi.com/apidojo/api/shazam. From there I changed it into a parser to suit my own needs.

#This is a program that lets you input any word and outputs a bunch of songs with a related name. i.e if you put in
#kiss you get "kiss me thru the phone".

def get_autofill(key_word):
    autofill = requests.request("GET", 
                                'https://shazam.p.rapidapi.com/auto-complete',
                                headers= headers,
                                params= {"term":f"{key_word}","locale":"en-US"}).text
    print(autofill)

#this is a program that lets you input the name of an artist and outputs their artist id. when making the program below
#this one, i realized that artist id was required to search for an artists library. Therefore, i made a program that 
#searched for the artist id using the search endpoint so that we can use that program
    
def get_artist_id(artist):
    artists = requests.request("GET", 
                                "https://shazam.p.rapidapi.com/search", 
                                headers= headers, 
                                params= {"term":f"{artist}","locale":"en-US","offset":"0","limit":"5"}).json()
    adamid = artists['tracks']['hits'][1]['track']['artists'][0]['adamid']
    print(adamid)

#this program lets you input the artist id you can find above and outputs a list of that artists top songs    
    
def get_artist_top_songs(artist_id):
    response = requests.request("GET", 
                                "https://shazam.p.rapidapi.com/artists/get-top-songs", 
                                headers=headers, 
                                params={"id":artist_id,"l":"en-US"}).json()
    data = response['data']
    
    for i in data:
        name = i['attributes']['name']
        print(name)

get_artist_id("Rihanna")
get_artist_top_songs("63346553")
```
