```python
import requests
from tkinter import *
import random

#the presentation looks kinda ugly but it's functional

root = Tk()
root.geometry("500x500")

author_list = []
slug_list = []

def get_author_list():
    author = requests.get(f"https://api.quotable.io/random")
    content_author = author.json()
    random_author = content_author['author']
    slug = content_author['authorSlug']

    author_list.append(random_author)
    slug_list.append(slug)

    author = requests.get(f"https://api.quotable.io/random")
    content_author = author.json()
    random_author = content_author['author']
    slug = content_author['authorSlug']

    author_list.append(random_author)
    slug_list.append(slug)
    
    author = requests.get(f"https://api.quotable.io/random")
    content_author = author.json()
    random_author = content_author['author']
    slug = content_author['authorSlug']

    author_list.append(random_author)
    slug_list.append(slug)
    
    author = requests.get(f"https://api.quotable.io/random")
    content_author = author.json()
    random_author = content_author['author']
    slug = content_author['authorSlug']

    author_list.append(random_author)
    slug_list.append(slug)

    author = requests.get(f"https://api.quotable.io/random")
    content_author = author.json()
    random_author = content_author['author']
    slug = content_author['authorSlug']

    author_list.append(random_author)
    slug_list.append(slug)

def clear_author_list():
    author_list.clear()
    slug_list.clear()

def get_authors():
    
    clear_author_list()
    get_author_list()
    
    author_box.delete(0.0, END)
    author_box2.delete(0.0, END)
    author_box3.delete(0.0, END)
    author_box4.delete(0.0, END)
    author_box5.delete(0.0, END)
    author_box.insert(END, author_list[0])
    author_box2.insert(END, author_list[1])
    author_box3.insert(END, author_list[2])
    author_box4.insert(END, author_list[3])
    author_box5.insert(END, author_list[4])

def get_random_quotes(slug):
    quotes = requests.get(f"https://api.quotable.io/quotes?author={slug}")
    content_quotes = quotes.json()
    results = content_quotes['results']

    quote_list = []

    for i in results:
        quotes = i['content']
        quote_list.append(quotes)

    random_quote = random.choice(quote_list)

    return random_quote
    
def get_quote1():
    quote1 = get_random_quotes(slug_list[0])
    quote_box.delete(0.0, END) 
    quote_box.insert(END, quote1)

def get_quote2():
    quote2 = get_random_quotes(slug_list[1])
    quote_box.delete(0.0, END) 
    quote_box.insert(END, quote2)

def get_quote3():
    quote3 = get_random_quotes(slug_list[2])
    quote_box.delete(0.0, END) 
    quote_box.insert(END, quote3)
    
def get_quote4():
    quote4 = get_random_quotes(slug_list[3])
    quote_box.delete(0.0, END) 
    quote_box.insert(END, quote4)
    
def get_quote5():
    quote5 = get_random_quotes(slug_list[4])
    quote_box.delete(0.0, END) 
    quote_box.insert(END, quote5)
        
label = Label(root, text="Authors")
label.pack()

new_list = Button(root, text="Click to get a new set of authors", command=get_authors)
new_list.pack()

author_box = Text(root, height=2, width=15, padx=20, pady=15)
author_box.pack()

quote1 = Button(root, text="Click here to get a quote from this author", command=get_quote1)
quote1.pack()

author_box2 = Text(root, height=2, width=15, padx=20, pady=15)
author_box2.pack()

quote2 = Button(root, text="Click here to get a quote from this author", command=get_quote2)
quote2.pack()

author_box3 = Text(root, height=2, width=15, padx=20, pady=15)
author_box3.pack()

quote3 = Button(root, text="Click here to get a quote from this author", command=get_quote3)
quote3.pack()

author_box4 = Text(root, height=2, width=15, padx=20, pady=15)
author_box4.pack()

quote4 = Button(root, text="Click here to get a quote from this author", command=get_quote4)
quote4.pack()

author_box5 = Text(root, height=2, width=15, padx=20, pady=15)
author_box5.pack()

quote5 = Button(root, text="Click here to get a quote from this author", command=get_quote5)
quote5.pack()

quote_box = Text(root, font="Elephant", height=4, width=30, bg="light blue", padx=30, pady=50)
quote_box.pack()

quote = requests.get(f"https://api.quotable.io/random")
content_quote = quote.json()
random_author = content_quote['author']
random_author

root.mainloop()

#I completely forgot to implement the drop down menu, and at this point I am so deep into this code that I don't know how
#to add it. I do know that I have to use OptionMenu(root, value, *item_list), and if I had more time I'd implement it.
```

```python
root = Tk()
root.title("7 day forecast")

lat = "42.8864"
lon = "-78.8784"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
weather_json = genius.json()
forecast = weather_json["properties"]["forecast"]
new_request = requests.get(forecast)
json_1 = new_request.json()
full = json_1["properties"]["periods"]

periods = []
temps = []
details = []

for i in full:
    periods.append(i["name"])
    temps.append(i["temperature"])
    details.append(i["detailedForecast"])
    
for i in range(14):
    time = Label(root, text = str(periods[i]), pady = 15, padx = 15, font = ("Times New Roman", 10))
    time.grid(row=i, column=0)
    temp = Label(root, text=str(temps[i]), font = ("Times New Roman", 10))
    temp.grid(row=i, column=2)
    detail = Label(root, text=str(details[i]), font = ("Times New Roman", 10))
    detail.grid(row=i, column=4)

root.mainloop()
```
