```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
import csv

csv_file = open("employment_data.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Job", "Type", "Application Deadline", "Salary"])

PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://www.binghamton-ny.gov/home")

government = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/a")

departments = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/div/div/ul[1]/li/a")

actions = ActionChains(driver)
actions.click(government)
actions.click(departments)
actions.perform()

try:
    civil = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='widget_4_33_127']/ul/li[16]/a"))
    )
    civil.click()
    
    employment = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='leftNav_1038_0_145']/ul/li/ul/li[14]/ul/li/a"))
    )
    employment.click()
    
    job_openings = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='ColumnUserControl4']/div[2]"))
    )
    
    table = job_openings.find_elements(By.TAG_NAME, "tr")
    
    for i in table:
        job = i.find_element(By.TAG_NAME, "a").text

        typ = i.find_element(By.XPATH, "//*[@id='ColumnUserControl4']/div[2]/table/tbody").text

        deadline = i.find_element(By.CLASS_NAME, "mobile_hide").text

        salary = i.find_element(By.XPATH, "//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr/td[4]").text
        print(job, typ, deadline, salary)

        csv_writer.writerow([job, typ, deadline, salary])
        
finally:
    driver.quit()
    
csv_file.close()

#when I tried to put a . in front of the xpaths above I ran into an error.
#I noticed that this could be because each row has a different xpath. Unfortunately, I had no idea how to fix this.
#I also noticed that most of these columns don't have classes, but instead "data-th" next to it. This confused me
#because I didn't know how else to locate the element in the column. When I tried to use xpaths I ran into the
#aforementioned problem.
```
