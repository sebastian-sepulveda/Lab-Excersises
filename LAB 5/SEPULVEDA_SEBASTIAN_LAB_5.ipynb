```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re

PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

practice_site = "https://techwithtim.net"
driver.get(practice_site)

search = driver.find_element(By.NAME, "s")
search.clear()
search.send_keys("test")
search.send_keys(Keys.RETURN)

try:
    main = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "main"))
    )
    print(main.text)
    
    articles = main.find_elements(By.TAG_NAME, "article")
    
    for i in articles:
        header = i.find_element(By.CLASS_NAME, "entry-summary")
        print(header.text)
finally:
    driver.quit()
```

```python
###### TUTORIAL 3 ######

PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://techwithtim.net") #establishing a link to a new website 

link = driver.find_element(By.LINK_TEXT,"Python Programming") #locates the link tag with this specific text attached to it
link.click() #this clicks on the link established in the line above

try: 
    element = WebDriverWait(driver, 10).until(   #making the driver wait on the next order so that it doesn't crash
    EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials")) #locates the link for Beginner Python Tutorials
    )
    element.click() #clicks on the link to Beginner Python Tutorials to open that page
    
    element = WebDriverWait(driver, 10).until(    #waiting again so that the page has time to load in
    EC.presence_of_element_located((By.ID, "sow-button-19310003")) #locates the button with the ID that has this specific name
    )
    element.click()  #clicks on the "Get Started" button to open up that page
    
    driver.back() #driver.back() goes back once to the previous page
    driver.back()
    driver.back()
    driver.forward() #driver.forward() returns to the next page we were at before you went back
    driver.forward()
except:
    driver.quit() #exits out of the page
```

```python
#scraping hoopshype
#use Selenium to open up the page salaries > players & scrape the players salaries. Store the data in a csv file 

import csv

csv_file = open("player_salaries.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Player", "Salary 2022/23"])

PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver2 = webdriver.Chrome(service=PATH)

driver2.get("https://hoopshype.com/")

link = driver2.find_element(By.LINK_TEXT,"SALARIES")
link.click()

try:
    element = WebDriverWait(driver2, 10).until(
    EC.presence_of_element_located((By.LINK_TEXT, "PLAYER SALARIES"))
    )
    element.click()
    
    main = WebDriverWait(driver2, 10).until(
    EC.presence_of_element_located((By.ID, "content"))
    )
    
    table = main.find_elements(By.TAG_NAME, "tr")
    
    for i in table:
        players = i.find_element(By.CLASS_NAME, "name")
        print(players.text)
        
        year2022 = i.find_element(By.CLASS_NAME, "hh-salaries-sorted")
        print(year2022.text)
        
        csv_writer.writerow([players.text, year2022.text])

finally:
    driver2.quit()

csv_file.close()
```
